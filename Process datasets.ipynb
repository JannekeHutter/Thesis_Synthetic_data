### Script to process BN datasets to correct format
Run script in thesis_venv.
Put target column at the end; & all columns (also categorical ones) are integer columns

```python
import pandas as pd
import json
from sklearn import preprocessing
```

```python
dataset_name = "Sachs2" # "Sachs" # "Asia" # "Cancer" #  "Cancer2" # 
real_file_root = "../BN_Datasets/" #"../Real_Datasets/" #
real_path = real_file_root + dataset_name+"/"+dataset_name
with open(real_path+"_meta.json") as json_file:
    meta = json.load(json_file)
categorical_cols = meta["categorical_cols"]
print("categorical_cols:", categorical_cols)

real_path = real_path +"_cat.csv"
real_data = pd.read_csv(real_path)
#categorical_cols = ["Pollution", "Smoker", "Cancer", "Pollution2", "Smoker2", "Cancer2"]
#categorical_cols = ["asia", "tub", "smoke", "either", "xray", "dysp"]
real_data.head()
```

```python
# Manual labeling for Sachs data:
df = real_data
for c in categorical_cols:
    df.loc[df[c] == "HIGH", c] = 2
    df.loc[df[c] == "AVG", c] = 1
    df.loc[df[c] == "LOW", c] = 0
df.describe()
```

```python
# Target Akt is the only binary col
df.loc[df["Akt"] == "HIGH", "Akt"] = 1
df.loc[df["Akt"] == "LOW", "Akt"] = 0
df.describe()
```

```python
# Manual labeling for Cancer data:
df = real_data
df.loc[df["Pollution"] == "low", "Pollution"] = 0
df.loc[df["Pollution"] == "high", "Pollution"] = 1
df.loc[df["Smoker"] == False, "Smoker"] = 0
df.loc[df["Smoker"] == True, "Smoker"] = 1
df.loc[df["Cancer"] == False, "Cancer"] = 0
df.loc[df["Cancer"] == True, "Cancer"] = 1
df.loc[df["Pollution2"] == "low", "Pollution2"] = 0
df.loc[df["Pollution2"] == "high", "Pollution2"] = 1
df.loc[df["Smoker2"] == False, "Smoker2"] = 0
df.loc[df["Smoker2"] == True, "Smoker2"] = 1
df.loc[df["Cancer2"] == False, "Cancer2"] = 0
df.loc[df["Cancer2"] == True, "Cancer2"] = 1
```

```python
df.describe()
```

General label encoder (check if it picks up intuitive encoding (yes/positive/True = 1, etc), otherwise encode manually as shown above)

```python
real_ints = real_data
for c in categorical_cols:
    print("Processing column:", c, "\n")
    # make new LabelEncoder for each column:
    le = preprocessing.LabelEncoder()
    real_ints[c]=le.fit_transform(real_data[c])
```
